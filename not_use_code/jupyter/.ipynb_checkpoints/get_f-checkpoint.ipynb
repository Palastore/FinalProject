```python
class MyPrettyPrinter(pprint.PrettyPrinter):
    def format(self, object, context, maxlevels, level):
        if isinstance(object, unicode):
            return (object.encode('thai'), True, False)
        return pprint.PrettyPrinter.format(self, object, context, maxlevels, level)

def go_print( input ):
    MyPrettyPrinter().pprint(input)
    # ppp = pprint.PrettyPrinter(indent=4)
    # ppp.pprint(input)
    return;

def get_midnight(time):
    return time.replace(minute=0, hour=0, second=0, microsecond=0)

def get_thai_midnight(time):
    out = time + datetime.timedelta(hours=7)
    out = out.replace(minute=0, hour=0, second=0, microsecond=0) - datetime.timedelta(hours=7)
    return out
```

```python
if __name__ == '__main__':

    client = MongoClient()
    db = client[db_name]

    cursor = db[col_name].find().sort([('ts', pymongo.ASCENDING)])

    date = []

    date_h = []

    bar = progressbar.ProgressBar(maxval=cursor.count()+1, widgets=[progressbar.Bar('#', '[', ']'), ' ', progressbar.Percentage()])
    bar.start()
    data=1
    for doc in cursor:
        data += 1
        bar.update(data)

        ts = doc['ts']
        datetime_object = datetime.datetime.fromtimestamp(ts/1000)
        name_day = doc['created_at'].split()[0]

        if not doc['hashtags']:
            date_h.append((datetime_object,name_day))

        date.append((datetime_object,name_day))
    bar.finish()
```

```python
def two_str(num):
    if num < 10:
        return str(0)+str(num)
    else:
        return str(num)
```

```python
def count_f(date,time_gap):
    all_name_day = {}
    all_min = {}

    ex_day = -1
    count_day = -1
    ex_hour = -1
    ex_min = -1
    count_min = -1
    for datetime_object,day in date:
        m = datetime_object.minute
        h = datetime_object.hour
        if day != ex_day:
            if count_day != -1:
                if not all_name_day.has_key(ex_day):
                    all_name_day[ex_day] = []
                all_name_day[ex_day].append(count_day)
            count_day = 0
            ex_day = day
        if ex_hour != h or ex_min != m:
            if ex_min%time_gap == 0:
                if count_min != -1:
                    key = two_str(ex_hour)+'.'+two_str(ex_min)
                    if not all_min.has_key(key):
                        all_min[key] = []
                    all_min[key].append(count_min)
                count_min = 0
            ex_hour = h
            ex_min = m
        count_day += 1
        count_min += 1

    if count_day != -1:
        if not all_name_day.has_key(ex_day):
            all_name_day[ex_day] = []
        all_name_day[ex_day].append(count_day)

    if count_min != -1:
        while (ex_min%time_gap != 0):
            ex_min += 1
            if ex_min == 60:
                ex_min = 0
                ex_hour += 1
                if ex_hour == 24:
                    exhout = 0
        key = key = two_str(ex_hour)+'.'+two_str(ex_min)
        if not all_min.has_key(key):
            all_min[key] = []
        all_min[key].append(count_min)

    return all_name_day,all_min
```

```python
def get_avg(dic):
    out = {}
    for k,v in dic.items():
        out[k] = (numpy.mean(v),len(v))
    return out
```

```python
def plot(doc,title):
    bar_x = []
    bar_y = []
    for x,y in sorted(doc.items()):
        bar_x.append(x)
        bar_y.append(y[0])
    pylab.plot(bar_x, bar_y)
    pylab.title(title)
```
